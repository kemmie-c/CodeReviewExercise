```python
import pandas as pd
import re

# ===== 讀取 Excel =====
excel_file_path = '/content/train_Data.xlsx'
df = pd.read_excel(excel_file_path)
df.head()


from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = df['content_cleaned']
y = df['label']

# Split the data into training and testing sets with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=100,  # 100 samples for the test set
    stratify=y,     # Stratify based on the 'label' column
    random_state=42 # for reproducibility
)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")

print("\nDistribution of labels in y_train:")
display(y_train.value_counts())

print("\nDistribution of labels in y_test:")
display(y_test.value_counts())


import jieba

def segment_text(text):
    if isinstance(text, str):
        seg_list = jieba.cut(text, cut_all=False)
        return " ".join(seg_list)
    else:
        return ""

# Apply word segmentation to the cleaned content in both training and testing sets
X_train_segmented = X_train.apply(segment_text)
X_test_segmented = X_test.apply(segment_text)

print("Original X_train sample:")
display(X_train.head())

print("\nSegmented X_train sample:")
display(X_train_segmented.head())

print("\nOriginal X_test sample:")
display(X_test.head())

print("\nSegmented X_test sample:")
display(X_test_segmented.head())


# ========= 中文停用詞（保留關鍵詞，不會把你指定的廣告字當停用詞）=========
chinese_stopwords = [
    "的", "了", "是", "在", "我", "有", "和", "也", "就", "與", "及",
    "不", "但", "而", "或", "被", "一個", "沒有", "你", "他", "她",
    "我們", "你們", "他們", "因此", "這個", "那個", "如果", "因為",
    "所以", "此外", "而且", "此外", "仍然", "並且"
]

# ========= 改進 1: 使用 TF-IDF + 停用詞 =========
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    min_df=2,
    max_features=5000,
    stop_words=chinese_stopwords
)

X_train_bow = vectorizer.fit_transform(X_train_segmented)
X_test_bow = vectorizer.transform(X_test_segmented)

print("Shape of X_train_bow:", X_train_bow.shape)
print("Shape of X_test_bow:", X_test_bow.shape)


# ============ KW boosting  (boost_factor = 120.0) ==========
# 這些字詞將被放大，使模型對它們更敏感，但不強制模型輸出某 label
key_phrases = [
    # sample list - 可視需要擴充或精簡
    "有效", "快速", "立即", "迅速", "立刻", "超強", "超級", "強效", "特強",
    "極速", "極佳", "最佳", "完美", "卓越", "顯著", "明顯",
    "安全", "可靠", "無效", "天然", "純天然", "高效",
    "專家推薦", "醫生推薦", "權威推薦",
    "熱銷", "限量", "特價", "促銷", "優惠",
    "補充", "改善", "舒緩", "保養", "修復", "強化", "提升", "促進",
    "健康", "營養", "有助", "抗氧化", "瘦身", "減重",
    "療效", "治療", "保健", "功能", "功效"
]

boost_factor = 120.0  # 你指定的值

# 我們使用 scipy.sparse 的 CSC 格式安全地將指定欄位的 data 放大
import scipy.sparse as sp

def boost_keywords(sparse_matrix, vectorizer, keywords, boost_factor=120.0):
    """
    對 sparse_matrix (CSR 或其他) 中，位於 vectorizer 詞彙表的 keywords 欄位進行放大。
    返回放大後的 CSR 矩陣（同類型）。
    """
    # 轉成 CSC 以便操作每個欄(column)的 data slice（更有效率）
    if not sp.isspmatrix_csc(sparse_matrix):
        csc = sparse_matrix.tocsc()
    else:
        csc = sparse_matrix.copy()

    vocab = vectorizer.vocabulary_  # dict: token -> col_index
    for kw in keywords:
        if kw in vocab:
            j = vocab[kw]
            start = csc.indptr[j]
            end = csc.indptr[j+1]
            if end > start:
                # multiply the column slice in-place
                csc.data[start:end] = csc.data[start:end] * boost_factor
    # 返回 CSR（原先程式使用 CSR 的常見 API）
    return csc.tocsr()

# 應用 boosting（不改變原變數名稱）
X_train_bow = boost_keywords(X_train_bow, vectorizer, key_phrases, boost_factor=boost_factor)
X_test_bow = boost_keywords(X_test_bow, vectorizer, key_phrases, boost_factor=boost_factor)

# 檢查（簡單顯示被放大的某些欄位 sum）
vocab = vectorizer.vocabulary_
for kw in key_phrases[:10]:
    if kw in vocab:
        idx = vocab[kw]
        # 計算該欄在 training set 的總和（可選）
        col_sum = X_train_bow[:, idx].sum()
        print(f"Keyword '{kw}' exists in vocab at index {idx}, col_sum(train) = {col_sum:.4f}")

# ========== End boosting ==========


import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping


# ========= 改進 2: 增加隱藏層 + Dropout =========
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_bow.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()


# Train the model
y_train_numerical = y_train.astype(int)
y_test_numerical = y_test.astype(int)

# 轉成 dense（注意：可能吃記憶體，若資料過大請改用小 batch 或其他策略）
X_train_bow_dense = X_train_bow.todense()
X_test_bow_dense = X_test_bow.todense()

early_stop = EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)

history = model.fit(
    X_train_bow_dense, y_train_numerical,
    epochs=10,
    shuffle=True,
    batch_size=64,
    validation_split=0.3,
    callbacks=[early_stop]
)

# ================================================================
# ===================== CHINESE FONTS =========================

!wget -O taipei_sans_tc_beta.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download
!mv taipei_sans_tc_beta.ttf /usr/local/lib/python3.12/dist-packages/matplotlib/mpl-data/fonts/ttf/

import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont('/usr/local/lib/python3.12/dist-packages/matplotlib/mpl-data/fonts/ttf/taipei_sans_tc_beta.ttf')
mpl.rc('font', family='Taipei Sans TC Beta')

# ================== CHINESE FONTS END ========================


# =========================================================
# ===================== SHAP BLOCK =========================
# ========== 放在訓練完成後、模型評估前 ======================
# =========================================================

import shap
import numpy as np

print("\n=== SHAP: computing global explanations ===")

# background sample：取 training data 前 300 筆（若不足則取全部）
n_background = min(300, X_train_bow_dense.shape[0])
background = np.array(X_train_bow_dense[:n_background])

# DeepExplainer 適用於 Keras 深度模型
explainer = shap.DeepExplainer(model, background)

# 注意：DeepExplainer 回傳的 shap_values 在二元 sigmoid 輸出時通常是 list(len=1) 的 array
shap_values = explainer.shap_values(np.array(X_test_bow_dense))

# ---- Summary plot ----
# 若 shap_values 是 list，取第 0 個輸出
sv = shap_values[0] if isinstance(shap_values, list) else shap_values
# Explicitly reshape sv to ensure it's 2D (samples, features)
sv = sv.reshape(sv.shape[0], -1)

# Convert X_test_bow_dense to a numpy array explicitly for SHAP plotting functions
X_test_features_for_shap = np.array(X_test_bow_dense)

shap.summary_plot(
    sv,
    X_test_features_for_shap,
    feature_names=vectorizer.get_feature_names_out()
)

# ---- Bar plot ----
shap.summary_plot(
    sv,
    X_test_features_for_shap,
    feature_names=vectorizer.get_feature_names_out(),
    plot_type="bar"
)

# ---- Dependence plot ----
top_features = vectorizer.get_feature_names_out()
# 預設選第一個 feature（你也可以改成其他 feature name）
feature_to_plot = top_features[0]
feature_index = int(list(top_features).index(feature_to_plot))

print(f"\nDrawing dependence plot for feature: {feature_to_plot}")

shap.dependence_plot(
    feature_index,
    sv,
    X_test_features_for_shap,
    feature_names=vectorizer.get_feature_names_out()
)

# ---- Force plot ----
# 在 notebook/Colab 中可互動式顯示
shap.initjs()

sample_id = 0
print(f"\nGenerated SHAP force plot for test sample #{sample_id}")

# explainer.expected_value 可能是 array/list，取第一個元素
expected_val = explainer.expected_value
if isinstance(expected_val, (list, np.ndarray)):
    expected_val = expected_val[0]

shap.force_plot(
    expected_val,
    sv[sample_id],
    X_test_features_for_shap[sample_id],
    feature_names=vectorizer.get_feature_names_out()
)

# ================== SHAP BLOCK END ========================


# ============= SHAP FEATURE (+),(+) CONTRIBUTION ==================

import pandas as pd
import numpy as np

# === 計算每個詞對「預測 1」與「預測 0」的 SHAP 貢獻強度 ===

shap_matrix = sv   # (samples, features)
feature_names = vectorizer.get_feature_names_out()

# 正向 SHAP（推向 1）的貢獻強度（取正值）
pos_strength = np.sum(np.where(shap_matrix > 0, shap_matrix, 0), axis=0)

# 負向 SHAP（推向 0）的貢獻強度（取負值的絕對值，越大代表越推向 0）
neg_strength = np.sum(np.where(shap_matrix < 0, -shap_matrix, 0), axis=0)

# 避免除以 0
total_strength = pos_strength + neg_strength
total_strength = np.where(total_strength == 0, 1, total_strength)

# 百分比
pos_ratio = pos_strength / total_strength
neg_ratio = neg_strength / total_strength

# 做成 DataFrame
shap_strength_df = pd.DataFrame({
    "feature": feature_names,
    "positive_strength": pos_strength,
    "negative_strength": neg_strength,
    "positive_ratio": pos_ratio,
    "negative_ratio": neg_ratio
})

# 排序：最常推模型成 1 的詞彙
top_pos = shap_strength_df.sort_values("positive_strength", ascending=False)
# 排序：最常推模型成 0 的詞彙
top_neg = shap_strength_df.sort_values("negative_strength", ascending=False)

display(top_pos.head(30))
display(top_neg.head(30))



import matplotlib.pyplot as plt

# === Top 30 正向 SHAP 強度 ===
plt.figure(figsize=(10, 12))
plt.barh(top_pos.head(30)["feature"], top_pos.head(30)["positive_strength"])
plt.xlabel("Positive SHAP Strength (pushing toward label=1)")
plt.title("Top 30 Words Driving Prediction Toward Class 1")
plt.gca().invert_yaxis()
plt.show()

# === Top 30 負向 SHAP 強度 ===
plt.figure(figsize=(10, 12))
plt.barh(top_neg.head(30)["feature"], top_neg.head(30)["negative_strength"])
plt.xlabel("Negative SHAP Strength (pushing toward label=0)")
plt.title("Top 30 Words Driving Prediction Toward Class 0")
plt.gca().invert_yaxis()
plt.show()


# ============= SHAP FEATURE (+),(+) CONTRIBUTION END ==================


###模型評估
# X_test_bow_dense 已在上面預先建立
loss, accuracy = model.evaluate(X_test_features_for_shap, y_test_numerical)
print(f"\nTest Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")


# Convert y_test to numerical dtype if it hasn't been already
y_true = y_test.values.astype('float32')
display("y_true =", y_true)

y_pred_prob = model.predict(X_test_features_for_shap).ravel().astype('float32')
display("y_pred_prob =", y_pred_prob)


# =========================================================
# ================= ROC, Confusion Matrix =================
# =========================================================

from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

fpr, tpr, _ = roc_curve(y_true[:], y_pred_prob[:])
roc_auc = auc(fpr, tpr)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)

ax.plot(fpr, tpr, lw=2, label=f'AUC = {roc_auc:.2f}')
ax.plot([0,1], [0,1], 'k--', lw=2)
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.legend(loc="lower right")
plt.show()


threshold = 0.55
y_pred = y_pred_prob > threshold
display("y_true =", y_true.astype(int))
display("y_pred =", y_pred.astype(int))

from sklearn.metrics import confusion_matrix
import seaborn as sns

y_pred_prob2 = model.predict(X_test_features_for_shap)
y_pred2 = (y_pred_prob2 >= 0.55).astype(int).flatten()
cm = confusion_matrix(y_test_numerical, y_pred2)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['True 0', 'True 1'])
plt.title('Confusion Matrix - Test Set')
plt.show()


# 自訂 cost function
def cost_fn(y_true, y_pred_prob, threshold=0.55):
    [[tn, fp], [fn, tp]] = confusion_matrix(y_true, y_pred_prob > threshold)
    cost = (tp * -10000) + (fn * -200000) + (fp * -10000) + (tn * 200000)
    return int(cost)

cost_fn(y_true, y_pred_prob2, threshold=0.55)


from sklearn.metrics import classification_report
print("Classification Report:")
print(classification_report(y_test_numerical, y_pred2))


import numpy as np
X_test_dense = X_test_bow.toarray()
y_test_numerical = y_test.astype(int).values

y_pred_prob = model.predict(X_test_dense)
y_pred = (y_pred_prob >= 0.7).astype(int).flatten()

results = pd.DataFrame({
    "text": X_test,
    "true": y_test_numerical,
    "pred": y_pred
})

tn = results[(results.true == 0) & (results.pred == 0)]
tp = results[(results.true == 1) & (results.pred == 1)]
fn = results[(results.true == 1) & (results.pred == 0)]
fp = results[(results.true == 0) & (results.pred == 1)]

examples = []
if len(tn) > 0: examples.append(("True Negative", tn.sample(1)))
if len(tp) > 0: examples.append(("True Positive", tp.sample(1)))
if len(fn) > 0: examples.append(("False Negative", fn.sample(1)))
if len(fp) > 0: examples.append(("False Positive", fp.sample(1)))

if len(examples) < 5:
    more_samples = results.sample(5 - len(examples))
    for _, row in more_samples.iterrows():
        examples.append(("Random", row.to_frame().T))

for label, row in examples:
    print(f"\n[{label}]")
    print("Text:", row["text"].values[0])
    print("True Label:", row["true"].values[0])
    print("Pred Label:", row["pred"].values[0])


##預測資料
predict_df = pd.read_csv("/content/predict_sample.csv")

predict_df["content_segmented"] = predict_df["content_cleaned"].apply(segment_text)

X_predict_bow = vectorizer.transform(predict_df["content_segmented"])
X_predict_dense = X_predict_bow.todense()

y_proba = model.predict(X_predict_dense).ravel()
y_pred = (y_proba >= 0.55).astype(int)

output_df = pd.DataFrame({
    "ID": predict_df["ID"],
    "label": y_pred
})

output_file = "group2_陳萱鎂_answer_v0.csv"
output_df.to_csv(output_file, index=False, encoding="utf-8-sig")
output_df
```
